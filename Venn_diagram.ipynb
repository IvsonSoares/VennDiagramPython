```python
%matplotlib inline
import matplotlib_venn as vplt
from matplotlib_venn import venn3
from matplotlib_venn import venn3_circles
from matplotlib import pyplot as plt
```

<p>Definicao dos conjuntos A e B</p>
<p>A = {0,2,4,6,8}</p>
<p>B = {2,3,5,7,9}</p>

```python
#Conjuntos A e B
A = {0,2,4,6,8}
B = {2,3,5,7,9}
#Universo para poder calcular o complemento
U = set([i for i in range(0,16)])
```

a) A $ \cup $ B

```python
C = A.union(B)
D = {'N'}

plt.figure(figsize=(10,10))
plt.title("Uniao de A e B: " + str(C))
v = vplt.venn2([A,B], set_labels = ('A', 'B'))
c = vplt.venn2_circles(subsets = (A, B));
#Text
v.get_label_by_id('100').set_text('\n'.join(map(str,A-(A&B))))
v.get_label_by_id('110').set_text('\n'.join(map(str,A&B)))
v.get_label_by_id('010').set_text('\n'.join(map(str,B-(A&B))))
#Color
#A
v.get_patch_by_id('100').set_color('blue')
v.get_patch_by_id('100').set_alpha(0.8)
#Intersec
v.get_patch_by_id('110').set_color('blue')
v.get_patch_by_id('110').set_alpha(0.8)
#B
v.get_patch_by_id('010').set_color('blue')
v.get_patch_by_id('010').set_alpha(0.8)
#universe
plt.annotate('U', xy=(0.7,0.5), xytext=(20,-20), ha='center', textcoords='offset points', fontsize=20)
plt.axis('on')
plt.show()
```

b) $ \overline {A  \cup  B}$

```python
C = A.union(B)
D = {'N'}

plt.figure(figsize=(10,10))
plt.title("Complemento da uniao de A e B ")
v = vplt.venn2([A,B], set_labels = ('A', 'B'))
c = vplt.venn2_circles(subsets = (A, B));
#Text
v.get_label_by_id('100').set_text('\n'.join(map(str,A-(A&B))))
v.get_label_by_id('110').set_text('\n'.join(map(str,A&B)))
v.get_label_by_id('010').set_text('\n'.join(map(str,B-(A&B))))
#Color
#A
v.get_patch_by_id('100').set_color('white')
v.get_patch_by_id('100').set_alpha(0.8)
#Intersec
v.get_patch_by_id('110').set_color('white')
v.get_patch_by_id('110').set_alpha(0.8)
#B
v.get_patch_by_id('010').set_color('white')
v.get_patch_by_id('010').set_alpha(0.8)
#background 'universe'
plt.annotate('U', xy=(0.7,0.5), xytext=(20,-20), ha='center', textcoords='offset points', fontsize=20)
plt.gca().set_facecolor('blue')
plt.gca().set_axis_on()
plt.axis('on')
plt.show()
```
